## **Dataset Description**

#### URL: https://www.kaggle.com/datasets/unsdsn/world-happiness


The dataset analyzed in this literate programming document looks at the global happiness levels of 155 countries from 2015, 2016, 2017, 2018 and 2019. The data in this report was collected from the Gallup World Poll and released at an United Nations event celebrating International Day of Happiness on March 20th (Sustainable Development Solutions Network, 2019). This dataset can be found on Kaggle posted for public use by the Sustainable Development Solutions. 

The main piece of data in the World Happiness Report is the happiness rating which is on a scale 0 to 10 where 10 is the best possible life for the respondent and 0 being the worst. This report also has data on GDP per Capita, family, life expectancy, freedom, government corruption and trust, generosity, and dystopia residual. Most of the variables are self-explanatory however, the Sustainable Development Solutions explains in the description of the dataset on Kaggle that the dystopia residual variable or the "unexplained components, differ for each country, reflecting the extent to which the six variables either over- or under-explain average 2014-2016 life evaluations" (Sustainable Development Solutions Network, 2019). They also explain that the Dystopia Residual metric formula is the Dystopia Happiness Score (1.85) + the unexplained Residual value.
